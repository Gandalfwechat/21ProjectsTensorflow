```python
# This is  a practice to apply RL Q learning  

import numpy as np 
import matplotlib.pyplot as plt


EnvStaStrs ='''
..........
.        .
.    x   .
.  x  o  .
.        .
..........
'''

def getMap():
    EnvStaSpace = [] 
    for line in EnvStaStrs.split('\n'):
        for c in line: 
            EnvStaSpace.append(c)  
    return EnvStaSpace
EnvStaSpaceUse= getMap()
```

```python
class env(object): 
    
    def __init__(self):
        self.Ax=1
        self.Ay=1 
        self.length =10
        self.width = 6
        self.endGame = False 
        self.MoveAction= [[0,1],[1,0],[0,-1],[-1,0]] 
    #                    down  lef   up     sright    
        self.totalReward =0
        
    def Get_State(self,x,y):
        return self.length*y+ x 
    def Interact(self, a):
        if ~self.endGame: 
            tryX = self.Ax + self.MoveAction[a][0]
            tryY = self.Ay + self.MoveAction[a][1] 
            stateID = self.Get_State(tryX,tryY) 
            if EnvStaSpaceUse[stateID]=='.':
                reward = 0
            if EnvStaSpaceUse[stateID]=='o':
                reward = 100 
                self.updateloc(tryX,tryY)
                self.endGame = True
            if EnvStaSpaceUse[stateID] == ' ': 
                reward = 0 
                self.updateloc(tryX,tryY) 
            if EnvStaSpaceUse[stateID] == 'x':                
                reward = -50 
                self.endGame = True
            self.totalReward+=reward
        return reward 
        
    def printMap(self):
        stateID = self.Get_State(self.Ax,self.Ay)   
        strOut = ''
        EnvStaSpacePrint = getMap()
        EnvStaSpacePrint[stateID]='A'
        for i in range(self.width):
            line =''
            for j in range(self.length): 
                staId = i*self.length+ j 
                line+=EnvStaSpacePrint[staId]
            line+='\n'
            strOut+=line
        print(strOut)
    def updateloc(self,x,y):     
        self.Ax = x
        self.Ay = y

        
# to do is to add update function     
```

```python

Alpha=0.9
gamma =0.7 
e=env()
Q = np.zeros([e.length*e.width,4])

def epsilon_Greedy(Q, state,eps):
    epstemp = np.random.rand(1)
    if epstemp[0]>eps:
        a = np.random.randint(0,4)
    else: 
        a = np.argmax(Q[state,:])
    return a

stepsList = []
for i in range(300):
   # print('------------------- ',str(i),' episod')
    e=env()
    step=0
    while not e.endGame and step<50:
        state=e.Get_State(e.Ax,e.Ay)
        
        a=epsilon_Greedy(Q,state,0.5)
        
        reward = e.Interact(a)
        
        newSta = e.Get_State(e.Ax,e.Ay)
        
        Q[state,a] = (1-Alpha)*Q[state,a]+Alpha*(reward + gamma*np.max(Q[newSta,:]))
        
        step+=1
     #   print(e.totalReward)
        e.printMap()
    stepsList.append(step)

plt.plot(stepsList)
```

```python
# let the agent play the game
# according to the Q table
stepsListPlay = []
for i in range(5):

    e=env()
    step=0
    while not e.endGame and step<50:
        state=e.Get_State(e.Ax,e.Ay)
        
        a=epsilon_Greedy(Q,state,1)
        
        reward = e.Interact(a)
        
        newSta = e.Get_State(e.Ax,e.Ay)
         
        step+=1
      #  print(e.totalReward)
        e.printMap()
    if e.endGame:
        strVictory='Got gold！'
    else:
        strVictory=''
    print(str(i+1),' episod',strVictory,' use ',step, 'steps')
    stepsListPlay.append(step)
```
